```python
# %% [markdown]
# # Snake AI - Træning & Analyse
# 
# Notebook til at køre eksperimenter, analysere resultater og visualisere modeller.

# %% Setup
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import json

from configs.config import get_config, list_experiments
from train import Trainer, run_experiment
from nn_model import NeuralNetwork
from snake_game import SnakeGame
from plot_utils import plot_single_experiment, plot_comparison, generate_all_plots
from visualize import watch, compare_models

# %% Se tilgængelige eksperimenter
list_experiments()


# ## 1. Kør et enkelt eksperiment

# %% Baseline træning
trainer = Trainer("franken_v4")

# Optional: Live plotting callback
def live_plot(trainer, gen):
    if gen % 10 == 0:
        clear_output(wait=True)
        fig, axes = plt.subplots(1, 3, figsize=(15, 4))
        
        # Fitness
        axes[0].plot(trainer.metrics.generations, trainer.metrics.best_fitness, label='Best')
        axes[0].plot(trainer.metrics.generations, trainer.metrics.avg_fitness, label='Avg', alpha=0.7)
        axes[0].set_title("Fitness")
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        
        # Score
        axes[1].plot(trainer.metrics.generations, trainer.metrics.best_score, label='Best', color='orange')
        axes[1].plot(trainer.metrics.generations, trainer.metrics.avg_score, label='Avg', color='yellow', alpha=0.7)
        axes[1].set_title("Score (mad spist)")
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)
        
        # Diversity
        axes[2].plot(trainer.metrics.generations, trainer.metrics.diversity, color='purple')
        axes[2].set_title("Diversity")
        axes[2].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()

trainer.on_generation = live_plot
metrics = trainer.train(print_every=10, save_every=25)


# ## 2. Kør flere eksperimenter

# %% Kør alle eksperimenter
experiments = ["franken_v4", "baseline_v2", "larger_pop_v2", "larger_nn_v2", "franken"]

for exp_name in experiments:
    print(f"\n{'='*50}")
    print(f"Kører: {exp_name}")
    run_experiment(exp_name, print_every=20)


# ## 3. Generer og sammenlign plots

# %% Generer plots for alle eksperimenter
generate_all_plots(show=False)

# %% Sammenlign eksperimenter (inline)
plot_comparison(["franken_v3", "franken_v4" "larger_nn_v2", "baseline_v2", "larger_pop_v2"], show=True)


# ## 4. Detaljeret analyse af ét eksperiment

# %% Load og analyser metrics
def analyze_experiment(experiment_name: str):
    """Detaljeret analyse af et eksperiment."""
    with open(f"results/{experiment_name}/metrics.json") as f:
        data = json.load(f)
    
    print(f" Analyse: {experiment_name}")
    print("=" * 40)
    
    # Fitness
    print(f"\n Fitness:")
    print(f"   Peak:  {max(data['best_fitness']):,.0f} (gen {np.argmax(data['best_fitness']) + 1})")
    print(f"   Final: {data['best_fitness'][-1]:,.0f}")
    print(f"   Avg final: {data['avg_fitness'][-1]:,.1f}")
    
    # Score
    if "best_score" in data:
        print(f"\n Score (mad spist):")
        print(f"   Max:  {max(data['best_score'])}")
        print(f"   Final best: {data['best_score'][-1]}")
        print(f"   Avg final: {data['avg_score'][-1]:.2f}")
    
    # Diversity
    print(f"\n Diversity:")
    print(f"   Start: {data['diversity'][0]:.3f}")
    print(f"   Final: {data['diversity'][-1]:.3f}")
    print(f"   Change: {((data['diversity'][-1] / data['diversity'][0]) - 1) * 100:.1f}%")
    
    # Death reasons
    if "deaths_wall" in data:
        total_deaths = (sum(data['deaths_wall']) + sum(data['deaths_tail']) + 
                       sum(data['deaths_loop']) + sum(data['deaths_max_moves']))
        print(f"\n Death reasons (total):")
        print(f"   Wall:  {sum(data['deaths_wall']):,} ({sum(data['deaths_wall'])/total_deaths*100:.1f}%)")
        print(f"   Tail:  {sum(data['deaths_tail']):,} ({sum(data['deaths_tail'])/total_deaths*100:.1f}%)")
        print(f"   Loop:  {sum(data['deaths_loop']):,} ({sum(data['deaths_loop'])/total_deaths*100:.1f}%)")
        print(f"   Max moves: {sum(data['deaths_max_moves']):,} ({sum(data['deaths_max_moves'])/total_deaths*100:.1f}%)")

analyze_experiment("franken_v3")

# %% [markdown]
# ## 5. Test model performance

# %% Test model med detaljeret output
def test_model_detailed(experiment_name: str, runs: int = 10):
    """Test model og vis detaljerede resultater."""
    config = get_config(experiment_name)
    weights = np.load(f"results/{experiment_name}/best_snake.npy")
    
    nn = NeuralNetwork.from_config(config, weights)
    game = SnakeGame(
        xsize=config["grid_size"],
        ysize=config["grid_size"],
        visual=False
    )
    
    results = []
    for i in range(runs):
        result = game.run_headless_detailed(
            nn, 
            max_moves=config["max_moves"],
            fitness_config=config["fitness"]
        )
        results.append(result)
        print(f"  Run {i+1:2d}: Score={result.score:2d}, Moves={result.moves:3d}, Death={result.death_reason}")
    
    scores = [r.score for r in results]
    moves = [r.moves for r in results]
    
    print(f"\n {experiment_name} - {runs} runs:")
    print(f"   Avg Score: {np.mean(scores):.1f}")
    print(f"   Max Score: {max(scores)}")
    print(f"   Avg Moves: {np.mean(moves):.0f}")
    
    # Death distribution
    deaths = {}
    for r in results:
        deaths[r.death_reason] = deaths.get(r.death_reason, 0) + 1
    print(f"   Deaths: {deaths}")

test_model_detailed("baseline", runs=10)

# Kør denne celle for at se AI'en spille live
watch("franken_v4", speed=10)

```
